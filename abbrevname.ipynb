# Scraping

```python
#Dependency
import pandas as pd
```

```python
#URL to scrape table from
url = 'http://www.sportsforecaster.com/nhl/abbreviations'
```

```python
#pull up list of tables
tables = pd.read_html(url)
tables
```

```python
#Get table with abbreviations and area/state name
city_abbrev = city_abbrev = tables[3]
city_abbrev.head(5)
```

```python
#rename columns for more clear site
city_abbrev = city_abbrev.rename(columns={0: 'team', 1: 'city'})
city_abbrev.head()
```

```python
#Format the 'team' column to uppercase
city_abbrev['team'] = city_abbrev['team'].str.upper()
city_abbrev.head()
```
