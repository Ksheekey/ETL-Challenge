# API CALL 1

```python
#doing an API call on google places to find the lat and lang of the city
#using the city name as the starting point

# Dependencies
import requests
import json
from pprint import pprint
from config import gkey
import time

city_ltlg = []
for city in city_list:
    target_city = city
    #print(target_city)

    # Build the endpoint URL
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
        'address={0}&key={1}').format(target_city, gkey)
    geo_data = requests.get(target_url).json()
    lat = geo_data["results"][0]["geometry"]['bounds']['northeast']["lat"]
    lng = geo_data["results"][0]["geometry"]["bounds"]['northeast']["lng"]
    #print(f"{lat}, {lng}")
    

    city_ltlg.append({"city":target_city,
                     "city lat_lng":f"{lat}, {lng}"})

print(city_ltlg)
```

```python
#putting the list of dictionaries into a df and dropping duplicates to merge
dict_df = pd.DataFrame(city_ltlg)
dict_df_dd = dict_df.drop_duplicates(subset=['city'])
dict_df_dd
```

```python
#merging the lat/lang info into the database
PTALL_df = pd.merge(PTA_df,dict_df_dd, on='city', how='left')
PTALL_df
```

# API CALL 2

```python
#using the lat and lang found in the previous API call to now find the closest restaurant to that city
restaurant_name = []
for cord in PTALL_df['city lat_lng']:
    # setting the geocoordinates to find the stadium
    target_coordinates = cord
    target_radius = 50000
    target_type = "restaurant"

    # set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "radius": target_radius,
        "type": target_type,
        "key": gkey
    }

    # base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)
    places_data = response.json()
    res_name = places_data['results'][0]['name']
    
    restaurant_name.append({"city lat_lng":cord,
                           "restaurant": res_name})
    #print(restaurant_name)
    #break
    
print(restaurant_name)
```

```python
#putting that list into a df to merge and dropping duplicates
restaurant_df = pd.DataFrame(restaurant_name)
restaurant_df_dd = restaurant_df.drop_duplicates(subset=['restaurant'])
restaurant_df_dd
```

```python
final_df = pd.merge(PTALL_df,restaurant_df_dd, on='city lat_lng', how='left')
del final_df['teamname']
final_df
```
