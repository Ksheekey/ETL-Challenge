```python
faceoff_data_df = pd.read_csv(faceoff_data)
```

```python
phi_win_data = faceoff_data_df.loc[(faceoff_data_df["team"] == "PHI") & (
    faceoff_data_df["win"] == True), :]
#phi_win_data
```

```python
phi_win_data = phi_win_data[['player','team','win']]
#phi_win_data
```
