```python
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_National_Hockey_League_arenas'
```

```python
tables = pd.read_html(url, header=[0])
tables
```

```python
df = df = tables[0]
df.head(20)
```

```python
df['Team'] = df['Team(s)'].astype(str).str.split()
```

```python
my_info = []
for info in df['Team']:
    my_info.append(info)
    
my_info
```

```python
#abc = pd.DataFrame(df['Team'])
headers = my_info.pop(0)
team_name = pd.DataFrame(my_info, columns=headers)
team_name
```

```python
list(team_name.columns)
```

```python
team_name = team_name.rename(columns={'Tampa': 'column1', 'Bay': 'column2', 'Lightning': 'column3'})
team_name
```

```python
team_name['team_name'] = team_name['column2'] + " " + team_name['column3']
team_name
```

```python
name_of_team = team_name[['column1','column2']]
name_of_team
```

```python
team_name__ = team_name[team_name['team_name'].notna()]
team_name_ = team_name__[['column1','team_name']]
team_name_
```

```python
toronto = team_name_.loc[team_name_['column1'] == 'Toronto']
```

```python
name_of_team.update(toronto)
name_of_team
```
