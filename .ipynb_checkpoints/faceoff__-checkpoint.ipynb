# READING FROM CSV

```python
# Dependencies
import pandas as pd
```

```python
# Store csv filepath in a variable
#kaggle was used to find the csv
#https://www.kaggle.com/garyrogers/nhl20192020faceoffdata
faceoff_data = "Resources/faceoff_data.csv"
faceoff_data_df = pd.read_csv(faceoff_data)
```

```python
#display table
faceoff_data_df
```

```python
#grab just the columns wanted
faceoff_data_df = faceoff_data_df[['player','team','win']]
faceoff_data_df
```

```python
#find only the faceoff wins
win_data = faceoff_data_df.loc[(faceoff_data_df["win"] == True), :]
win_data
```

```python
#group by player and team
grouped_team_info = win_data.groupby(['player','team'])
grouped_team_info
```

```python
#group by the total wins so there is no repeat players and reset the index
faceoff_win_data = grouped_team_info[['win']].sum()
faceoffs__ = faceoff_win_data.reset_index()
faceoffs__
```

```python
#sort the win column so the player with the most faceoffs wins is on top
faceoff_sort = faceoffs__.sort_values('win', ascending=False)
faceoff_sort = faceoff_sort.reset_index(drop=True)
faceoff_sort.head(50)
```

```python
#only pull the ones with more than 30 faceoff wins
over_30_wins = faceoff_sort.loc[(faceoff_sort["win"] > 30), :]
over_30_wins
```

```python
#import the city abbreviation table to merge with above
import import_ipynb as ipynb
from abbrevname import city_abbrev
city_abbrev
```

```python
#merge the two tables
team_info = pd.merge(over_30_wins,city_abbrev, on='team',how='left')
team_info
```
