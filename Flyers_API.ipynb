```python
# Dependencies
import requests
import json
from pprint import pprint
from config import gkey

#putting a target city to find its coordinates
#for city in city_df__:
target_city = 'Philadelphia'

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_city, gkey)
```

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(target_url).json()

# Print the json
pprint(geo_data)
```

```python
# Extracting the latitude and longitude to plug into nearby search
lat = geo_data["results"][0]["geometry"]['bounds']['northeast']["lat"]
lng = geo_data["results"][0]["geometry"]["bounds"]['northeast']["lng"]
```

```python
# setting the geocoordinates to find the stadium
target_coordinates = f"{lat}, {lng}"
target_radius = 50000
target_type = "stadium"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
pprint(places_data)
```

```python
# Print the name and address of the first restaurant that appears
flyers_stadium = places_data["results"][1]["name"]
```
